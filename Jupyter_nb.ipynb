Basic Operator::  "+"
the plus operator produces a new tuple, list or string

```python
tup1 = (23,34,'df') # tuple
tup2 = (34,532,4,'sd','kj')
tup3 = tup1+tup2
print(tup3)
```

```python
list1 = ['we','wwds',34,9,'lk'] # list
list2 = ['sdf','gfdc',34,98,3]
list3 = list1 + list2
print(list3)
```

```python
str1 = 'refdc' # str
str2 = 'gbfvdc'
str3 = str1+str2
print(str3)
```
